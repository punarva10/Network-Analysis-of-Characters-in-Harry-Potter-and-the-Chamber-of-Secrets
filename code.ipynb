```python
import nltk
import pandas as pd
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
from nltk.tokenize import sent_tokenize
```

```python
with open('Harry_Potter_and_the_chamber_of_secrets.txt', 'r') as f:
    contents = f.read()
```

```python
print(contents)
```

```python
characters = ['harry potter', 'hermione granger', 'ron weasley', 'tom riddle', 'gilderoy lockhart','ginny weasley', 'dumbledore ', 'hagrid', 'severus snape', 'lucious malfoy', 'draco malfoy', 'dobby' ]
```

```python
print(len(characters))
```

```python
import nltk
nltk.download('punkt')
sentences = sent_tokenize(contents)
print(len(sentences))
```

```python
print(sentences[0:10])
```

```python
print(sentences[0])
```

```python
contents = contents.replace('Harry','Harry Potter')
contents = contents.replace('Potter','Harry Potter')
contents = contents.replace('the boy who lived', 'Harry Potter')
contents = contents.replace('Scarhead', 'Harry Potter')
contents = contents.replace('Gryffindor Seeker', 'Harry Potter')

contents = contents.replace('Hermione','Hermione Granger')
contents = contents.replace('Granger','Hermione Granger')
contents = contents.replace('Miss Granger','Hermione Granger')

contents = contents.replace('Ron','Ron Weasley')
contents = contents.replace('Weasley','Ron Weasley')
contents = contents.replace('Ronnie','Ron Weasley')
contents = contents.replace('Ronnikins','Ron Weasley')
contents = contents.replace('Weasel','Ron Weasley')

contents = contents.replace('Tom Marvolo Riddle','Tom Riddle')
contents = contents.replace('TOM MARVOLO RIDDLE','Tom Riddle')
contents = contents.replace('Lord Voldemort','Tom Riddle')
contents = contents.replace('LORD VOLDEMORT','Tom Riddle')
contents = contents.replace('He- Who-Must-Not-Be-Named','Tom Riddle')
contents = contents.replace('He-Who-Must-Not-Be-Named','Tom Riddle')
contents = contents.replace('You-Know-Who','Tom Riddle')


contents = contents.replace('Professor Lockhart','Gilderoy Lockhart')
contents = contents.replace('Lockhart','Gilderoy Lockhart')
contents = contents.replace('The Defense Against the Dark Arts teacher','Gilderoy Lockhart')

contents = contents.replace('Ginny','Ginny Weasley')

contents = contents.replace('Snape','Severus Snape')
contents = contents.replace('Professor Snape','Severus Snape')
contents = contents.replace('Potions master','Severus Snape')

contents = contents.replace('Mr. Malfoy','Lucious Malfoy')

contents = contents.replace('Draco','Draco Malfoy')
contents = contents.replace('ferret','Draco Malfoy')

contents = contents.replace('house-elf','Dobby')

#confusion about what to do with 'Malfoy'
```

```python
char_df = pd.DataFrame(0, index=characters, columns=characters)
char_df.head()
```

```python
for sentence in sentences:
    sentence = sentence.lower()
    for char1 in characters:
        char1 = char1.lower()
        for char2 in characters:
            char2 = char2.lower()
            if (char1 != char2) and char1 in sentence and char2 in sentence:
                char_df[char1][char2] += 1
                char_df[char2][char1] += 1
char_df.head()
```

```python
G = nx.Graph()
G = nx.from_pandas_adjacency(char_df)
```

```python
print("Number of nodes in the graph:", len(G.nodes()))
print("Nodes:", G.nodes())
print()
print("Number of edges in the graph:", len(G.edges()))
print("Edges:", G.edges(data=True))
```

# 1.c CENTRALITY MEASURES :

### DEGREE:

```python
from operator import itemgetter
degc = nx.degree_centrality(G)
sorted(degc.items(), key=itemgetter(1), reverse=True)
```

### BETWEENNESS 

```python
bet = nx.betweenness_centrality(G)
sorted(bet.items(), key=itemgetter(1), reverse=True)
```

### CLOSENESS

```python
clos = nx.closeness_centrality(G)
sorted(clos.items(), key=itemgetter(1), reverse=True)
```

### PAGE RANK 

```python
pgr=nx.pagerank(G)
sorted(pgr.items(), key=itemgetter(1), reverse=True)
```

```python
import pandas as pd
import seaborn as sns
centrality_measures = {
    'degree': degc,
    'betweenness': bet,
    'closeness': clos,
    'page rank':pgr
}
centrality = pd.DataFrame(centrality_measures)
centrality
```

HARRY POTTER HAS THE HIGHEST CENTRALITY , FOLLOWED BY RON AND HERMOINE , THIS IS IN ACCORDANCE WITH THE KNOWN FACT THAT HARRY , RON AND HERMONIE ARE THE LEAD/MAIN PROTAGONISTS OF THE BOOK .

```python
import collections
degree_sequence = sorted([d for n, d in G.degree()], reverse=True) 
degreeCount = collections.Counter(degree_sequence)
deg, cnt = zip(*degreeCount.items())

fig, ax = plt.subplots()
plt.bar(deg, cnt, width=0.80, color='purple')

plt.title("Degree Histogram")
plt.ylabel("Count")
plt.xlabel("Degree")
ax.set_xticks([d + 0.5 for d in deg])
ax.set_xticklabels(deg)

# draw graph in inset
plt.axes([0.4, 0.4, 0.5, 0.5])
A = (G.subgraph(c) for c in nx.connected_components(G))
# = list(A)[0]
Gcc = sorted(A, key=len, reverse=True)[0]
pos = nx.spring_layout(G)
plt.axis('off')
nx.draw_networkx_nodes(G, pos, node_size=20)
nx.draw_networkx_edges(G, pos, alpha=0.5)

plt.show()
```

### G(n,p) and G(n,m) generated graph-ERDOS RENYL MODEL:

### G(n,p):

```python
def graph(p):
    gr= nx.erdos_renyi_graph(12,p, seed=42)
    #pos = nx.circular_layout(gr)
    nx.draw(gr, node_color=range(12), node_size=400, cmap=plt.cm.Reds)
    plt.show()

graph(0.1)
graph(0.3)
graph(0.5)
graph(0.7)
```
